```python
## Import Data ## 
import pandas as pd

# ! options !
pd.set_option('display.max_columns', None)
pd.set_option('display.float_format', '{:.2f}'.format)


df_18 = pd.read_csv(
    '/home/nathan/fac/m2/s1/python/td1/valeursfoncieres-2018.txt', delimiter='|',nrows=100000)
df_19 = pd.read_csv(
    '/home/nathan/fac/m2/s1/python/td1/valeursfoncieres-2019.txt', delimiter='|',nrows=100000)
df_20 = pd.read_csv(
    '/home/nathan/fac/m2/s1/python/td1/valeursfoncieres-2020.txt', delimiter='|', nrows=100000)
df_21 = pd.read_csv(
    '/home/nathan/fac/m2/s1/python/td1/valeursfoncieres-2021.txt', delimiter='|', nrows=100000)

#df_18 = pd.read_csv('/Users/josephpelham/Desktop/SISE/Python/Donnees/valeursfoncieres-2018.txt', delimiter='|')
#df_19 = pd.read_csv('/Users/josephpelham/Desktop/SISE/Python/Donnees/valeursfoncieres-2019.txt', delimiter='|')
#df_20 = pd.read_csv('/Users/josephpelham/Desktop/SISE/Python/Donnees/valeursfoncieres-2020.txt', delimiter='|')
# df_21 = pd.read_csv('/Users/josephpelham/Desktop/SISE/Python/Donnees/valeursfoncieres-2021.txt', delimiter='|')
```

```python
dataframes = [df_18, df_19, df_20,df_21] 

# Remplacer les espaces par des _ dans colonnes
for df in dataframes:
    df.columns = df.columns.str.replace(' ', '_')

dataframes[0].head()
```

```python
## Change data types and remove useless variables 


columns_to_remove = ["Code_voie", "No_voie", "Voie", "Type_de_voie",
                    "1er_lot", "No_plan", "Code_departement", "Type_local", "Code_commune", "Section"]


for i in range(len(dataframes)): 
    # Créer une copie modifiée du DataFrame
    df_modified = dataframes[i].copy()

    # Delete observations with NaN Code_type_local & Valeur_fonciere
    df_modified = df_modified.dropna(subset=['Code_type_local'])
    df_modified = df_modified.dropna(subset=['Valeur_fonciere'])

    df_modified['Valeur_fonciere'] = df_modified['Valeur_fonciere'].str.replace(
        ',', '.').astype(float)

    df_modified['Code_postal'] = df_modified.groupby('Commune')['Code_postal'].transform(lambda x: x.fillna(
        method='ffill').fillna(method='bfill'))  # match les code postal vide avec les communes (bien)
    df_modified['Code_postal'] = df_modified['Code_postal'].astype(int)

    
    df_modified['No_voie'] = df_modified['No_voie'].apply(lambda x: str(int(x)) if pd.notnull(
        x) else '').astype(str)  # change le type si non Nan sinon affiche rien
    
    df_modified['Code_type_local'] = df_modified['Code_type_local'].astype(int)

    #  Créer la variable "adresse"
    df_modified['adresse'] = df_modified['No_voie'] + ' ' + df_modified['Type_de_voie'].astype(str) + ' ' + \
        df_modified['Code_voie'].astype(str) + ' ' + df_modified['Voie'].astype(str) + ' ' + \
        df_modified['Code_postal'].astype(str) + ' ' + df_modified['Commune']
    
    # Après la création de l'adresse on remove les variables concatenées et les autres inutiles
    df_modified = df_modified.drop(columns_to_remove, axis=1)
    
    #sauvegarde
    dataframes[i] = df_modified
```

```python

for i in range(len(dataframes)):
    # Créer une copie modifiée du DataFrame
    df_modified = dataframes[i].copy()
    
    # Remove duplicates (date_mutation et adresse)
    df_modified = df_modified.drop_duplicates(subset=["Date_mutation", "adresse"])
    # 'adresse' useless maintenant => remove
    df_modified = df_modified.drop("adresse", axis = 1)

    # sauvegarde
    dataframes[i] = df_modified
   
```

```python
# sample random 5% of each year
df_18_s = dataframes[0].sample(frac=.99, random_state=42)
df_19_s = dataframes[1].sample(frac=.99, random_state=42)
df_20_s = dataframes[2].sample(frac=.99, random_state=42)
df_21_s = dataframes[3].sample(frac=.99, random_state=42)

```

```python
# Combine sampled subsets
df = pd.concat([df_18_s, df_19_s, df_20_s, df_21_s], axis=0)
```

```python
# Remove columns with 75% NA

# Enlever les colonnes à la main ?? pas de pourcent => plus explicite

# % of NA
missing_perc = df.isna().sum() / len(df) * 100
# Define the threshold for dropping columns (75% in this case)
threshold = 75
# Use the dropna method to drop columns with more than 75% missing values
df = df.dropna(axis=1, thresh=len(df) * (1 - threshold / 100))
df.head()
```

```python
# Créer la variable "year"
df['Date_mutation'] = pd.to_datetime(df['Date_mutation'], format='%d/%m/%Y')

# Extract the year component and create a new 'year' column
df['year'] = df['Date_mutation'].dt.year
print(df.year.value_counts())
```
