```python
import pandas as pd
from importlib import reload
import functions
from functions import *
from warnings import filterwarnings
import plotly.express as px

reload(functions)
filterwarnings(action="ignore")
pd.set_option('display.max_columns', None)
```

```python
# vf2018= read_t("valeursfoncieres-2018.txt")
# vf2019= read_t("valeursfoncieres-2019.txt")
# vf2020= read_t("valeursfoncieres-2020.txt")
# vf2021= read_t("valeursfoncieres-2021.txt")

all= pd.read_csv("all.csv", index_col=0)
all.head()
```

```python
rep= all[["Code departement", "Commune", "Valeur fonciere"]].groupby(["Code departement", "Commune"], as_index=False).agg("sum")
#répartition de la valeur foncière par année
rep1= all[["mois", "jour", "annee", "Valeur fonciere"]].groupby(["annee", "mois", "jour"], as_index=False).agg("sum")
```

```python
#Les valeurs foncieres par annee et mois
fig = px.funnel(rep1, y='Valeur fonciere', x='mois', color='annee')
fig.show()
```

```python
fig= px.sunburst(rep1, path=['annee', 'mois', 'jour'], values='Valeur fonciere')
fig.show()
```

```python
#Le prix moyen par commune
rep2= all[["Commune", "Valeur fonciere"]].groupby("Commune", as_index=False).agg("mean")
fig = px.histogram(rep2, x="Commune", y="Valeur fonciere")
fig.show()
```

```python
#Le prix moyen par code postal
rep3= all[["Code postal", "Valeur fonciere"]].groupby("Code postal", as_index=False).agg("mean")
fig = px.pie(rep3, values='Valeur fonciere', names='Code postal')
fig.update_traces(textposition='inside')
fig.update_layout(uniformtext_minsize=12, uniformtext_mode='hide')
fig.show()
```

```python
#Le prix moyen par code departement
rep4= all[["Code departement", "Valeur fonciere"]].groupby("Code departement", as_index=False).agg("mean")
fig = px.line(rep4, x='Code departement', y="Valeur fonciere")
fig.show()
```

```python
#Le prix moyen par Section
rep5= all[["Section", "Valeur fonciere"]].groupby("Section", as_index=False).agg("mean")
fig = px.bar(rep5, x="Section", y="Valeur fonciere")
fig.show()
```

```python
rep6= all[["Nombre pieces principales", "Valeur fonciere"]].groupby("Nombre pieces principales", as_index=False).agg("mean")

#En moyenne les pièces coutent:

fig= px.sunburst(rep6, path=['Nombre pieces principales'], values='Valeur fonciere', title="Valeur foncière moyenne par nombre de pièce")
fig.show()

```
